# Formatação de PDF com FPDF2

Criação e formatação de pdf seguindo [este tutorial](https://www.youtube.com/playlist?list=PLjNQtX45f0dR9K2sMJ5ad9wVjqslNBIC0).

```python
from fpdf import FPDF
from fpdf.enums import XPos, YPos
```

```python
class PDF(FPDF):
    def header(self) -> None:
        _ = self.image('../../assets/logo_brain_facens_colorido.png',w=32, h=32, y=2)
        self.set_font('helvetica', 'B', 20)
        _ = self.cell(0, 10, 'Simulação de Cenário', new_x=XPos.LMARGIN, new_y=YPos.NEXT, align='C')
        self.ln(20)
        return super().header()
    
    def footer(self) -> None:
        self.set_y(-15)
        self.set_font('helvetica', 'I', 10)
        _ = self.cell(0, 10, f'Página {self.page_no()}/{{nb}}', align='C')
        return super().footer()
```

```python
file_name = 'test.pdf'
pdf = PDF('P', 'mm', 'A4')
pdf.alias_nb_pages()
```

```python
pdf.set_title('PDF Teste')
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

pdf.set_font('helvetica', '', 16)
for i in range(41):
    _ = pdf.cell(0, 10, f"Linha {i}", border=1, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
```

```python
_ = pdf.output(file_name)
print(f'Created {file_name}')
```
